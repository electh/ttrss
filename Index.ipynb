```python
"""
使用变量保存数据并进行加减乘除运算
"""
a = 321
b = 12
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
"""
使用type()检查变量的类型
"""
a = 100
b = 12.345
c = 1 + 5j
d = 'hello world'
e = True # 不能使用true、TRUE等

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
"""
使用print()函数输出带占位符的字符串
"""
a = int(input('a = '))
b = int(input('b = '))
print('%d + %d = %d' % (a, b, a + b))
print('%d - %d = %d' % (a, b, a - b))
print('%d * %d = %d' % (a, b, a * b))
print('%d / %d = %f' % (a, b, a / b))
print('%d // %d = %d' % (a, b, a // b)) # 整除
print('%d %% %d = %d' % (a, b, a % b)) # mod,求余数
print('%d ** %d = %d' % (a, b, a ** b)) # 指数
```

```python
"""
赋值运算符和符合赋值运算符
"""
a = 10
b = 3
a += b # a = a + b = 10 + 3 = 13
a *= (a+2) # a = a * (a + 2) = 13 * (13 + 2) = 195
print(a)
```

```python
"""
比较运算符和逻辑运算符的使用
"""

flag0 = 1 == 1            # True
flag1 = 3 > 2             # True
flag2 = 2 < 1             # False
flag3 = flag1 and flag2   # False
flag4 = flag1 or flag2    # True
flag5 = not(1 != 2)      # False
print('flag0 =', flag0)   # print函数输出多个值，以逗号分隔，输出默认以空格分开
print('flag1 =', flag1)
print('flag2 =', flag2)
print('flag3 =', flag3)
print('flag4 =', flag4)
print('flag5 =', flag5)
```

```python
"""
将华氏度转化为摄氏度
通过公式：C（摄氏度）=(F（华氏度）-32)÷1.8来转换。
"""
f = float(input('请输入华氏温度：'))
c = (f - 32) / 1.8
print('%.1f华氏度 = %.1f摄氏度' % (f, c))

print(f'{f:.1f}华氏度 = {c:.1f}摄氏度') # print(f'')中f表示对打印内容进行格式化
```

```python
"""
计算圆的周长和面积
"""
r = float(input('r = '))
c = 2 * 3.14 * r
s = 3.14 * r ** 2
print(f'r = {r:.1f} c = {c:.1f} s = {s:.1f}')
```
