# DAY1 语言元素

```python
"""
使用变量保存数据并进行加减乘除运算
"""
a = 321
b = 12
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
"""
使用type()检查变量的类型
"""
a = 100
b = 12.345
c = 1 + 5j
d = 'hello world'
e = True # 不能使用true、TRUE等

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
"""
使用print()函数输出带占位符的字符串
"""
a = int(input('a = '))
b = int(input('b = '))
print('%d + %d = %d' % (a, b, a + b))
print('%d - %d = %d' % (a, b, a - b))
print('%d * %d = %d' % (a, b, a * b))
print('%d / %d = %f' % (a, b, a / b))
print('%d // %d = %d' % (a, b, a // b)) # 整除
print('%d %% %d = %d' % (a, b, a % b)) # mod,求余数
print('%d ** %d = %d' % (a, b, a ** b)) # 指数
```

```python
"""
赋值运算符和符合赋值运算符
"""
a = 10
b = 3
a += b # a = a + b = 10 + 3 = 13
a *= (a+2) # a = a * (a + 2) = 13 * (13 + 2) = 195
print(a)
```

```python
"""
比较运算符和逻辑运算符的使用
"""

flag0 = 1 == 1            # True
flag1 = 3 > 2             # True
flag2 = 2 < 1             # False
flag3 = flag1 and flag2   # False
flag4 = flag1 or flag2    # True
flag5 = not(1 != 2)      # False
print('flag0 =', flag0)   # print函数输出多个值，以逗号分隔，输出默认以空格分开
print('flag1 =', flag1)
print('flag2 =', flag2)
print('flag3 =', flag3)
print('flag4 =', flag4)
print('flag5 =', flag5)
```

```python
"""
将华氏度转化为摄氏度
通过公式：C（摄氏度）=(F（华氏度）-32)÷1.8来转换。
"""
f = float(input('请输入华氏温度：'))
c = (f - 32) / 1.8
print('%.1f华氏度 = %.1f摄氏度' % (f, c))

print(f'{f:.1f}华氏度 = {c:.1f}摄氏度') # print(f'')中f表示对打印内容进行格式化
```

```python
"""
计算圆的周长和面积
"""
r = float(input('r = '))
c = 2 * 3.14 * r
s = 3.14 * r ** 2
print(f'r = {r:.1f} c = {c:.1f} s = {s:.1f}')
```

```python
"""
分段函数求值

        3x - 5  (x > 1)
f(x) =  x + 2   (-1 <= x <= 1)
        5x + 3  (x < -1)
"""
x = float(input('x = '))
if x > 1:
    y = 3 * x - 5
elif x >= -1:
    y = x + 2
else:
    y = 5 * x + 3
print(f'f({x:.1f}) = {y:.1f}')
```

```python
"""
分段函数求值，使用嵌套的分支结构（不建议使用复杂结构）

        3x - 5  (x > 1)
f(x) =  x + 2   (-1 <= x <= 1)
        5x + 3  (x < -1)
"""
x = float(input('x = '))
if x > 1:
    y = 3 * x - 5
else:
    if x >= -1:
        y = x + 2
    else:
        y = 5 * x + 3
print('f(%.1f) = %.1f' % (x, y))
```

```python
"""
百分制成绩转换为等级制成绩。
如果输入的成绩在90分以上（含90分）输出A；
80分-90分（不含90分）输出B；
70分-80分（不含80分）输出C；
60分-70分（不含70分）输出D；
60分以下输出E。
"""
score = float(input('score = '))
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'E'
print(f'对应的等级是{grade}')
```

# DAY3 循环结构

```python
"""
用for循环实现1~100求和
"""
sum = 0
for i in range(1, 101):
    sum += i
print(sum)
```

```python
"""
用for循环实现1~100之间的偶数求和
"""
sum = 0
for i in range(2, 101, 2):
    sum += i
print(sum)
```

```python
"""
用for循环实现1~100之间的偶数求和
另外一种实现方式（不建议）
"""
sum = 0
for i in range(1, 101):
    if i % 2 == 0:
        sum += i
    else:
        continue
print(sum)
```
